```python
import matplotlib.pyplot as plt
from scipy import stats

plt.rcParams["figure.figsize"] = (8, 6)

x = [10, 12, 15, 7, 14, 20, 25, 23, 25, 30, 25, 30, 23, 31, 26, 33, 19, 30, 25, 24, 27, 29, 9, 10, 11]
y = [15, 27, 19, 4, 27, 31, 38, 33, 53, 48, 51, 72, 56, 45, 40, 67, 22, 32, 31, 50, 55, 57, 6, 7, 8]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def estimate_lemonade_damend(x):
  return slope * x + intercept

estimated_lemonade_damends = list(map(estimate_lemonade_damend, x))

plt.scatter(x, y)
plt.plot(x, estimated_lemonade_damends)
plt.show()
print("slope", slope, "intercept", intercept, "correlation", r)
```

```python
train_x = x[:20]
train_y = y[:20]

test_x = x[20:]
test_y = y[20:]
```

```python
plt.scatter(train_x, train_y)
plt.show()
```

```python
plt.scatter(test_x, test_y)
plt.show()
```

Fit train set

```python
slope, intercept, r, p, std_err = stats.linregress(train_x, train_y)

def estimate_lemonade_damend(x):
  return slope * x + intercept

estimated_lemonade_damends = list(map(estimate_lemonade_damend, train_x))

plt.scatter(train_x, train_y)
plt.plot(train_x, estimated_lemonade_damends)
plt.show()
print("slope", slope, "intercept", intercept, "correlation", r)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(train_y, estimated_lemonade_damends)
print("R@-score (Determination)", r2)
```

test set

```python

estimted_lemonade_demands = list(map(estimate_lemonade_damend, test_x))

plt.scatter(test_x, test_y)
plt.plot(test_x, estimted_lemonade_demands)
plt.show()
```

```python
r2 = r2_score(test_y, estimted_lemonade_demands)
print("R@-score (Determination)", r2)
```
